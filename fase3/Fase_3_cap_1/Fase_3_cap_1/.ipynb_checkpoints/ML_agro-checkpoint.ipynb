```python
# --- CÉLULA 2: Importação das bibliotecas ---
import os
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report, confusion_matrix, f1_score, accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC

warnings.filterwarnings("ignore")
```

```python
# CÉLULA 3 – Leitura da Base de Dados

caminho_base = "arquivos/produtos_agricolas.csv"

df = pd.read_csv(caminho_base)

df.head()
```

```python
# CÉLULA 4 – Análise Exploratória dos Dados

plt.style.use("seaborn-v0_8")
plt.rcParams.update({
    "axes.titlesize": 14,
    "axes.labelsize": 12,
    "xtick.labelsize": 10,
    "ytick.labelsize": 10
})

plt.figure(figsize=(8,5))
sns.countplot(x="label", data=df, palette="viridis")
plt.title("Distribuição das Culturas (label)")
plt.xlabel("Tipo de Cultura")
plt.ylabel("Quantidade")
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.savefig("imagens/grafico_distribuicao.png")
plt.show()

plt.figure(figsize=(9,7))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="YlGnBu", fmt=".2f")
plt.title("Mapa de Correlação entre Variáveis")
plt.tight_layout()
plt.savefig("imagens/grafico_correlacao.png")
plt.show()

plt.figure(figsize=(8,6))
sns.scatterplot(x="temperature", y="humidity", hue="label", data=df, palette="tab10", s=50, edgecolor='k')
plt.title("Relação entre Temperatura e Umidade por Cultura")
plt.xlabel("Temperatura (°C)")
plt.ylabel("Umidade (%)")
plt.legend(title="Cultura", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig("imagens/grafico_temp_umid.png")
plt.show()

plt.figure(figsize=(8,6))
sns.boxplot(x="label", y="ph", data=df, palette="Set2")
plt.title("Distribuição do pH do Solo por Cultura")
plt.xlabel("Cultura")
plt.ylabel("pH do Solo")
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.savefig("imagens/grafico_ph.png")
plt.show()

plt.figure(figsize=(8,6))
sns.violinplot(x="label", y="rainfall", data=df, palette="coolwarm")
plt.title("Distribuição da Chuva por Cultura")
plt.xlabel("Cultura")
plt.ylabel("Chuva (mm)")
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.savefig("imagens/grafico_chuva.png")
plt.show()
```

```python
# Perfil ideal de solo e clima por cultura 

# médias por tipo de cultura (label)
perfil_ideal = df.groupby("label")[["N", "P", "K", "temperature", "humidity", "ph", "rainfall"]].mean().round(2)

# ordena pelo nome da cultura
perfil_ideal = perfil_ideal.sort_index()

print("📊 Perfil médio ideal de solo e clima por cultura:\n")
display(perfil_ideal)

# gráfico visual para comparação
plt.figure(figsize=(10,6))
sns.heatmap(perfil_ideal.T, cmap="YlGnBu", annot=True)
plt.title("Perfis Médios Ideais de Solo e Clima por Cultura")
plt.xlabel("Cultura")
plt.ylabel("Variável")
plt.tight_layout()
plt.savefig("imagens/perfil_ideal.png")
plt.show()
```

```python
# CÉLULA 5 – Preparação dos Dados para Machine Learning
# Nesta etapa, realizamos o pré-processamento da base de dados:
# Selecionamos as variáveis de entrada (X) e a saída (y)
# Convertendo os rótulos textuais em valores numéricos (LabelEncoder)
# Dividindo os dados em conjuntos de treino e teste (80/20)

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# seleciona as variáveis de entrada (X) e a saída (y)
X = df[["N", "P", "K", "temperature", "humidity", "ph", "rainfall"]]
y = df["label"]

# converte a saída (texto) para número
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# divide entre treino e teste (80% / 20%)
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)

print("✅ Dados prontos para treino!")
print("Treino:", X_train.shape, "Teste:", X_test.shape)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# inicializa os modelos
modelos = {
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "KNN": KNeighborsClassifier(n_neighbors=5),
    "SVM": SVC(kernel="rbf", gamma="scale"),
    "Logistic Regression": LogisticRegression(max_iter=1000)
}

# dicionário para armazenar os resultados
resultados = {}

for nome, modelo in modelos.items():
    modelo.fit(X_train, y_train)
    previsao = modelo.predict(X_test)
    acc = accuracy_score(y_test, previsao)
    resultados[nome] = round(acc * 100, 2)

# converte para DataFrame
import pandas as pd
resultado_df = pd.DataFrame(list(resultados.items()), columns=["Modelo", "Acurácia (%)"]).sort_values(by="Acurácia (%)", ascending=False)

print("📊 Comparativo de desempenho dos modelos:")
display(resultado_df)
```

```python
# --- CÉLULA 10: Avaliação detalhada do melhor modelo (Random Forest) ---

from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay

# Seleciona o melhor modelo
modelo_final = RandomForestClassifier(random_state=42)
modelo_final.fit(X_train, y_train)

# Faz previsões
y_pred = modelo_final.predict(X_test)

# Matriz de confusão
plt.figure(figsize=(8,6))
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=le.classes_)
disp.plot(cmap="Blues", values_format='d')
plt.title("Matriz de Confusão - Random Forest")
plt.tight_layout()
plt.savefig("imagens/matriz_confusao.png")
plt.show()

# Relatório de classificação
print("📋 Relatório de Classificação:\n")
print(classification_report(y_test, y_pred, target_names=le.classes_))
```
